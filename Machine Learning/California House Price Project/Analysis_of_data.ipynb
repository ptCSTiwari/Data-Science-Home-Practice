```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import  StratifiedShuffleSplit
```

```python
df = pd.read_csv("housing.csv")
```

```python
#here we are suntituting median income column with the income cat column beacuse it will help to classify data in more good manner
df['income_cat'] =pd.cut(df['median_income'], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels=[1,2,3,4,5])
```

```python
df["income_cat"].value_counts().sort_index().plot.bar(rot=0)  #rot is using only for changigng the x label workds direction
plt.title("Icome categries distribution")
plt.xlabel("income category")
plt.ylabel("No of salaries")
```

```python
def shuffle_split(data, test_ratio):
  np.random.seed(42)   #it will fix the random generated outputs
  shuffled_indices = np.random.permutation(len(data))  #it will suffle our data
  test_size = int(len(data)* test_ratio)
  test_indices = shuffled_indices[:test_size]
  train_indices = shuffled_indices[test_size:]
  return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train, test = shuffle_split(df, 0.2)
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['income_cat']):
  strata_train = df.loc[train_index]
  strata_test = df.loc[test_index]
```
