```python
from sklearn.tree import DecisionTreeClassifier
```

```python
features = [[150, 0],
[170, 0],
[130, 1],
[120, 1]]
```

```python
labels = ["apple", "apple", "orange", "orange"]
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(features, labels)
```

```python
prediction = clf.predict([[150, 1]])
```

```python
print(f"The fruit is {prediction[0]}")
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(11,6))
```

```python
plot_tree(clf, feature_names=["weight", "color"], class_names=["apple", "orange"], filled=True)
```

```python
plot_tree(clf, feature_names=["weight", "color"], class_names=["apple", "orange"], filled=False)
```
