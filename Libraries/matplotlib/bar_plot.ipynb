```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
sachin = [500, 700, 1100, 1800, 1500, 1200, 1323, 1309, 606, 1233]
kohli = [1300, 3453, 3410, 6443, 2434, 3434, 2343, 3543, 606, 1500]
rohit = [200, 450, 900, 1500, 3445, 4532, 1700, 1399, 243, 984]

x = np.arange(len(years))
width = 0.25

plt.figure(figsize=(12, 6))

# Bars
plt.bar(x - width, sachin, width=width, label="Sachin ramesh Tendulkar", color='#4A90E2')
plt.bar(x + width, kohli, width=width, label="Virat Kohli", color='#F5A623')
plt.bar(x, rohit, width=width, label="Rohit Sharma", color='#7ED321')

# Add styled labels
for i in range(len(years)):
    plt.text(x[i] - width, sachin[i] + 100, str(sachin[i]), ha='center', fontsize=8,
             fontweight='bold',
             bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.2', alpha=0.85))

    plt.text(x[i], rohit[i] + 100, str(rohit[i]), ha='center', fontsize=8,
             fontweight='bold',
             bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.2', alpha=0.85))

    plt.text(x[i] + width, kohli[i] + 100, str(kohli[i]), ha='center', fontsize=8,
             fontweight='bold',
             bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.2', alpha=0.85))

# Labels & title
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.barh(x - width, sachin, width=width, label="Sachin ramesh Tendulkar")
plt.barh(x + width, kohli, width=width, label="Virat Kohli")
plt.barh(x, rohit, width=width, label="Rohit Sharma")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
plt.show()
```
