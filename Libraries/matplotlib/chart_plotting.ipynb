```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
sachin =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, sachin)
plt.show()
```

```python
plt.plot(years, sachin)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
kohli = [1300, 3453, 3410, 6443, 2434, 3434, 1323, 3543, 606, 233]
plt.plot(years, sachin, label="Sachin ramesh Tendulkar")
plt.plot(years, kohli, label="Virat Kohli")
plt.xlabel("Year")
plt.ylabel("Scored runs")
plt.title("God vs King score comparasion")
# plt.legend() is used to display the legend on the plot.
# The legend helps identify which line corresponds to which label (e.g., "Sachin ramesh Tendulkar" or "Virat Kohli").
plt.legend()
plt.show()
```

```python
plt.plot(years, sachin, 'ro--', label="Sachin ramesh Tendulkar")
plt.plot(years, kohli, 'g3:', label="Virat Kohli")
```

```python
#decoration

kohli = [1300, 3453, 3410, 6443, 2434, 3434, 1323, 3543, 606, 233]
plt.plot(years, sachin, linewidth=3, label="Sachin ramesh Tendulkar")
plt.plot(years, kohli, linewidth=2, label="Virat Kohli")
plt.xlabel("Year")
plt.ylabel("Scored runs")
plt.title("God vs King score comparasion")
# plt.legend() is used to display the legend on the plot.
# The legend helps identify which line corresponds to which label (e.g., "Sachin ramesh Tendulkar" or "Virat Kohli").
plt.legend()
plt.show()
```
